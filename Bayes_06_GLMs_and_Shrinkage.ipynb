## Linear Regression
$
\begin{align*}
y_i &\sim{} \mathcal N\left(\mathbf{x}_i^\top \boldsymbol \beta, \sigma^2\right) && i = 1,\cdots,n\\
 &\;\quad {} \textrm{Assumptions} \\
1. &\;\quad {} \mathbf{x}_i^\top \boldsymbol \beta && \textrm{Linear model form true} && \hat y_i = \mathbf{x}_i^\top \boldsymbol{\hat \beta} \textrm{ from least squares}&& \hat \beta = \underset{\beta}{\textrm{argmin}} \sum_{i=1}^n(y_i - \hat y_i)^2 = \underset{\beta}{\textrm{argmin}} \; \hat{\textbf{e}}^\top\hat{\textbf{e}} \\
2. &\;\quad {} \sigma^2 && \textrm{Homoskedastity holds} && \sigma^2\neq\sigma_{x_i}^2 \textrm{ a function of } \mathbf{x}_i\\
3. &\;\quad {} \mathbf{x}_i && \textrm{Perfectly measured $\mathbf{x}_i$} && \mathbf{x}_i \textrm{ has no "uncertainty" }&&(\textit{$\mathbf{x}_i$ has no distributional assumptions})\\
4. &\;\quad {} \epsilon_i=y_i-\mathbf{x}_i^\top \boldsymbol \beta && \textrm{Normality of error terms} && \textrm{and residuals } \hat e_i=y_i-\hat y_i \\
\end{align*}$


## Generative Modeling

$p(y_i | \mathbf{x}_i, \boldsymbol {\hat \beta}, \hat \sigma)$ estimated as $\mathcal N\left(\mathbf{x}_i^\top \boldsymbol {\hat \beta}, \hat \sigma^2\right)$ is a **predictive model** as opposed to a **generative model**

Integrating out the estimation uncertainty provides the **posterior predictive distribution**

$p(y_i | \mathbf{x}_i) = \int p(y_i | \mathbf{x}_i, \boldsymbol {\hat \beta}, \hat \sigma) \; p(\boldsymbol {\hat \beta}, \hat \sigma)\; d \boldsymbol {\hat \beta} \; d \hat \sigma $

**Generative modeling** means considering stochastic rather than fixed $\mathbf{x}_i$

$p(y_i, \mathbf{x}_i) = p(y_i | \mathbf{x}_i) p(\mathbf{x}_i)  = \int p(y_i | \mathbf{x}_i, \boldsymbol \theta) \; p(\mathbf{x}_i | \boldsymbol \theta) \; p(\boldsymbol \theta)\; d \boldsymbol {\theta}$

## Least Squares

$
\begin{align*}
\hat \beta = \underset{\beta}{\textrm{argmin}} \sum_{i=1}^n(y_i - \hat y_i)^2 &={}
\underset{\beta}{\textrm{argmin}} \; \hat{\textbf{e}}^\top\hat{\textbf{e}}\\
&={} \underset{\beta}{\textrm{argmin}} \; (\textbf{y}-\textbf{X}\boldsymbol \beta)^\top(\textbf{y}-\textbf{X}\boldsymbol \beta)\\
&={} \underset{\beta}{\textrm{argmin}} \; \textbf{y}^\top \textbf{y} -2 (\textbf{X}\boldsymbol \beta)^\top \textbf{y} + (\textbf{X}\boldsymbol \beta)^\top (\textbf{X}\boldsymbol \beta)\\
\\
0 &={} \nabla_\beta \left[ \textbf{y}^\top \textbf{y} -2 \boldsymbol \beta^\top \textbf{X}^\top \textbf{y} + \boldsymbol \beta^\top \textbf{X}^\top \textbf{X}\boldsymbol \beta \right]\\
 &={} -2  \textbf{X}^\top \textbf{y} + 2 \textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} \\
 &={} - \textbf{X}^\top \textbf{y} + \textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} \\
 \\
\textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} &={} \textbf{X}^\top \textbf{y}  \\
\boldsymbol {\hat \beta} &={} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y}  \\
\hat{\textbf{y}} = \textbf{X} \boldsymbol {\hat \beta} &={} \textbf{X} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y} = \textbf{H} \textbf{y} = \hat{\textbf{y}} \\
\\
\textbf{X}^\top (\textbf{y}-\hat{\textbf{y}}) &={} \textbf{X}^\top \textbf{y} - \textbf{X}^\top\textbf{X} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y} = \textbf{X}^\top \textbf{y} - \textbf{X}^\top \textbf{y} = \textbf{0} \quad  \Longrightarrow{} \quad \sum_{i=1}^n \hat e_i = 0 \\
&\Longrightarrow{} \quad \textbf{X}^\top \textbf{y} =   \textbf{X}^\top \hat{\textbf{y}}
\end{align*}$

## Generalized Linear Models (GLM)

1. Not all data has **normally distributed error terms**
2. $E[Y|\textbf{x}_i] = \textbf{x}_i^T \boldsymbol \beta$ can paramaterize more than a **normal distribution**

$E[Y|\textbf{x}_i] = e^{\textbf{x}_i^T \boldsymbol \beta }$ parameterizes $\lambda > 0$ of a **Poisson distribution**

> A **Poisson** GLM uses a **log link function** so $\log(\lambda_i) = \textbf{x}_i^T \boldsymbol \beta$

$E[Y|\textbf{x}_i] = \frac{1}{1 + e^{-\textbf{x}_i^T \boldsymbol \beta }}$ parameterizes $p \in [0,1]$ of a **Bernoulli distribution**

> A **logistic regression** uses a **logit (log odds) link function** so $\frac{p}{1-p_i} = \textbf{x}_i^T \boldsymbol \beta$

An alternative is the **probit link function** $\Phi^{-1}(p_i) = \textbf{x}_i^T \boldsymbol \beta$ where $\Phi^{-1}$ is the inverse CDF of a standard normal distribution; thus, $E[y|\textbf{x}_i] = p_i = \Phi(\textbf{x}_i^T \boldsymbol \beta) = Pr(z_i \geq 0) \; \textrm{ for } \; z_i \sim \mathcal N(\textbf{x}_i^T \boldsymbol \beta,1)$

$
\begin{align*}
\Phi(\textbf{x}_i^T \boldsymbol \beta)\sqrt{2\pi} &={} \int_{-\infty}^{x_i^\top\boldsymbol{\beta}_{p\times 1}} e^{-\frac{1}{2}z^2_i}dz\\
&={} \int_{-\infty}^{0} e^{-\frac{1}{2}(z_i + x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz = \int_{-\infty}^{0} e^{-\frac{1}{2}(-z_i - x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz =  \int^{\infty}_{0} e^{-\frac{1}{2}(z_i-x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz
\end{align*}$

This formulation maps **binary** $y_i$ to a **latent normally distributed representation** $z_i$. This demonstrates a fundamental idea underlying **copulas** which transform arbitrary distribution into latent normal distributions (several of which) which may then be tied together as a multivariate normal distribution to provide inference on an underlying **covariance structure**.

- The general **copula** formulation for $j^{th}$ random variable $\textbf{Y}_j$ with CDF $F_{Y_j}$ is $\Phi^{-1}(F_{Y_j}(y_{ij}))$ but the trick used above is to instead map this as $\Phi^{-1}(F_{Y_j}(E[Y_{ij}|\textbf{x}_i|))$

## Shrinkage Regularization

**Machine Learning** fits models by optimizing penalized **loss functions**.

Two classic regularizations are "ridge" and "lasso" regression, which respectively use $L_2$ and $L_1$ penalty functions.

- Ridge: $$\sum_{i=1}^n \frac{1}{2}(y_i-x_i^T\beta_{p \times 1})^2 + \lambda \sum_{j=1}^n \beta_j^2 = \frac{1}{2}(y-X\beta)^T(y-X\beta) + \lambda \sum_{j=1}^n \beta_j^2 = \frac{1}{2}||y-X\beta||_2^2 + \lambda ||\beta||_2^2 $$
- Lasso: $$\sum_{i=1}^n \frac{1}{2}(y-x_i^T\beta_{p \times 1})^2 + \lambda \sum_{j=1}^n |\beta_j| = \frac{1}{2}(y-X\beta)^T(y-X\beta) + \lambda \sum_{j=1}^n |\beta_j| = \frac{1}{2}||y-X\beta||_2^2 + \lambda ||\beta||_1$$

### As priors

Bayes has been regularizing since [1763](https://en.wikipedia.org/wiki/Thomas_Bayes#Bayes'_theorem)

- **Machine Learning** is the original inventor or "regularization"?
- Bayesian analysis has been regularizing since it's inception: *priors are regularization*


**Ridge Regression** **$L_2$-penalty** is just a MAP estimate using **norml priors** for the **regression coefficients**

\begin{align*}
 \beta_i \sim {} & \text{Normal}(b_i,s_i) & f(\beta_i| b_i, s_i) = {}& \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{\beta_i-b_i}{s_i}\right)^2} & \overbrace{\underbrace{|\beta_i-b_i|^2}}_{\text{Squared Penalization}}^{L_2} \\
y_i \sim {} & \text{Normal}(x_i^T\beta, \sigma) & \sigma \sim {} & \text{HalfNormal}(\sigma_0)
\end{align*}

**Lasso Regression** **$L_1$-penalty** is just replacing the **normal priors** with **Laplace ("Double Exponential") prior distributions**

\begin{align*}
\beta_i \sim {} & \text{Laplace}(b_i,s_i) & f(\beta_i| b_i, s_i) = {}& {\frac  {1}{2b}}\exp \left(-{\frac{|\beta_i-b_i |}{s_i}}\right) & \overbrace{\underbrace{|\beta_i-b_i|}}_{\text{Absolute Penalization}}^{L_1}\\
y_i \sim {} & \text{Normal}(x_i^T\beta, \sigma) & \sigma \sim {} & \text{HalfNormal}(\sigma_0)
\end{align*}

We want to find $p(\sigma^2 | \mathbf{y}, \mathbf{X})$.

Likelihood:  
$$
p(\mathbf{y} | \sigma^2) \propto (\sigma^2)^{-n/2} \exp\left(-\frac{1}{2\sigma^2} (\mathbf{y} - \mathbf{X}\hat{\boldsymbol{\beta}})^\top (\mathbf{y} - \mathbf{X}\hat{\boldsymbol{\beta}})\right)
$$
where $\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top\mathbf{X})^{-1}\mathbf{X}^\top\mathbf{y}$ is the OLS estimate.
Prior:  
$$
p(\sigma^2) \propto (\sigma^2)^{-\alpha^* - 1} \exp\left(-\frac{\beta^*}{\sigma^2}\right)
$$


$$
p(\sigma^2 | \mathbf{y}, \mathbf{X}) \propto (\sigma^2)^{-n/2 - \alpha^* - 1} \exp\left(-\frac{1}{\sigma^2}\left[\beta^* + \frac{1}{2} (\mathbf{y} - \mathbf{X}\hat{\boldsymbol{\beta}})^\top (\mathbf{y} - \mathbf{X}\hat{\boldsymbol{\beta}})\right]\right)
$$

Let $n$ be the number of observations and $p$ the number of predictors.
The residual sum of squares is:  
$$
RSS = (\mathbf{y} - \mathbf{X}\hat{\boldsymbol{\beta}})^\top (\mathbf{y} - \mathbf{X}\hat{\boldsymbol{\beta}})
$$
The posterior becomes:  
$$
p(\sigma^2 | \mathbf{y}, \mathbf{X}) \propto (\sigma^2)^{-(\alpha^* + \frac{n}{2} + 1)} \exp\left(-\frac{1}{\sigma^2}\left[\beta^* + \frac{1}{2} \text{RSS}\right]\right)
$$

$\alpha_{\text{post}} = \alpha^* + \frac{n}{2}$  
$\beta_{\text{post}} = \beta^* + \frac{1}{2} \text{RSS}$

Thus:
$$
\sigma^2 | \mathbf{y}, \mathbf{X} \sim \text{Inverse-Gamma}\left(\alpha_{\text{post}}, \beta_{\text{post}}\right)
$$

```python
#Q3
import pymc as pm
import numpy as np
import pandas as pd
import arviz as az
import statsmodels.api as sm
import matplotlib.pyplot as plt

mtcars = sm.datasets.get_rdataset("mtcars").data.copy()
np.random.seed(365)
# Introduce outliers by dramatically lowering mpg for 3 random rows
lower_indices = np.random.choice(mtcars.index, size=3, replace=False)
mtcars.loc[lower_indices, 'mpg'] = mtcars.loc[lower_indices, 'mpg'] * 0.01

# Introduce outliers by dramatically increasing mpg for 3 random rows
# Ensure these indices are different from the lower_indices
remaining_indices = mtcars.index.difference(lower_indices)
higher_indices = np.random.choice(remaining_indices, size=3, replace=False)
mtcars.loc[higher_indices, 'mpg'] = mtcars.loc[higher_indices, 'mpg'] * 100


X = mtcars[['wt', 'hp']].values
X = np.hstack([np.ones((X.shape[0], 1)), X])  # add intercept
y = mtcars['mpg'].values
N, p = X.shape

with pm.Model() as robust_regression:
    beta = pm.Normal('beta', mu=0, sigma=10, shape=p)

    tau = pm.Gamma('tau', alpha=1, beta=1)

    nu = pm.Exponential('nu_minus_one', 1/30) + 1

    tau_i = pm.Gamma('tau_i', alpha=nu/2, beta=nu/2, shape=N)

    mu = pm.math.dot(X, beta)

    sigma = pm.Deterministic('sigma', 1/pm.math.sqrt(tau * tau_i))

    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed=y)

    trace = pm.sample(2000, tune=1000, target_accept=0.95, return_inferencedata=True)

az.summary(trace, var_names=["beta", "tau", "nu_minus_one"])
az.plot_trace(trace, var_names=["beta", "tau", "nu_minus_one"])
plt.show()

tau_post = trace.posterior['tau'].mean().values  # scalar
tau_i_post = trace.posterior['tau_i'].mean(dim=("chain", "draw")).values  # vector of length N
weights = tau_post * tau_i_post
D = np.diag(weights)

# Compute hat matrix using weighted least squares formula:
hat_matrix = X @ np.linalg.inv(X.T @ D @ X) @ (X.T @ D)
hat_diag = np.diag(hat_matrix)

# 7. Compare influence for outlier vs non-outlier points:
mtcars['hat_diag'] = hat_diag
mtcars['is_outlier'] = mtcars.index.isin(outlier_indices)
print(mtcars[['mpg', 'hat_diag', 'is_outlier']])

# Plot the influence measures:
lower_pos = np.where(mtcars.index.isin(lower_indices))[0]
higher_pos = np.where(mtcars.index.isin(higher_indices))[0]

plt.figure(figsize=(8, 4))
plt.stem(hat_diag, markerfmt='o', basefmt=" ", label="Influence (hat diag)")
plt.xlabel("Observation Index")
plt.ylabel("Hat Value")
plt.title("Posterior Influence Measures (Hat Diagonals)")

# Annotate lower mpg outliers in red
for pos in lower_pos:
    plt.annotate("Lower Outlier", (pos, hat_diag[pos]), color="red", fontsize=9, fontweight='bold')

# Annotate higher mpg outliers in blue
for pos in higher_pos:
    plt.annotate("Higher Outlier", (pos, hat_diag[pos]), color="blue", fontsize=9, fontweight='bold')

plt.legend()
plt.show()
```
