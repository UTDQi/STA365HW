### 3. What distribution do you think this $\theta$ has?

$\large
\begin{align*}
\prod_{i=1}^n\frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{1}{2\sigma^2}(x_i-\theta)^2} &={} (2\pi\sigma^2)^{-\frac{n}{2}} e^{-\frac{1}{2\sigma^2} \sum_{i=1}^n (x_i-\theta)^2}\\

\textrm{So} & \;\;\; \theta \sim N\left(\bar x, \frac{\sigma^2}{n}]\right)
\end{align*}

## Part C (III): Gibbs Sampling

The canonical example of ***Gibbs sampling*** is the ***normal-gamma*** model and here's a simplified version with $p(\theta,\tau) \propto 1$

$\begin{align*}
p\left(x|\theta,\tau=\frac{1}{\sigma^2}\right) = {} & \prod_{i=1}^n \sqrt{\frac{\tau}{2\pi}} e^{-\frac{\tau}{2}(x_i-\theta)^2}\\
p(\theta,\tau) \propto {} & 1 \\
p(\theta,\tau| x) \propto p(\theta,\tau, x) \propto {} &  \tau^{n/2} e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i-\theta)^2} \\
p(\tau|\theta, x) = {} & \text{Gamma}\left(\alpha = \frac{n}{2}+1, \textrm{rate} = \frac{1}{2} \sum_{i=1}^n (x_i-\theta)^2 \right)\\
p(\theta| \tau, x)  \propto {} & e^{-\frac{\tau}{2} \sum_{i=1}^n \theta^2 - 2\theta x_i} =  e^{-\frac{\tau}{2} \left(n\theta^2 - 2\theta \sum_{i=1}^n x_i\right)} = e^{-\frac{\tau n }{2} \left(\theta^2 - 2\theta \bar x\right)}\\
  \propto {} & e^{-\frac{n \tau}{2} (\theta - \bar x)^2} \\
p(\theta| \sigma^2 = \tau^{-1}, x) = {} & N\left(E[\theta] = \bar x, Var(\theta) = \frac{\sigma^2}{n}\right)
\end{align*}$

```python
#Q1

from scipy import stats
import numpy as np
import matplotlib.pyplot as plt


# loc is mean and scale is standard deviation
n = 30; x = stats.norm(loc=0, scale=1).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples
theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization

for c in range(C):
    for g in range(1, G):
        shape = n / 2
        scale = 2 / np.sum((x - theta[c, g-1]) ** 2)
        tau[c, g] = stats.gamma.rvs(a=shape, scale=scale)

        theta[c, g] = stats.norm.rvs(loc=np.mean(x), scale=np.sqrt(1 / (n * tau[c, g])))

burn = 20
demo = 120
posterior_theta = theta[:, burn:].flatten()
posterior_tau = tau[:, burn:].flatten()

interval_theta = np.percentile(posterior_theta, [2.5, 97.5])
interval_tau = np.percentile(posterior_tau, [2.5, 97.5])

print("95% CI for theta:", interval_theta)
print("95% CI for tau:", interval_tau)

fig,ax = plt.subplots(1, 4, figsize=(14,3)); ax[0].set_title("$\\theta$ Burn-In"); ax[1].set_title("$\\tau$ Burn-In"); ax[2].set_title("$\\theta$ Converged Markov Chain\nStationary Distribution Samples"); ax[3].set_title("$\\tau$ Converged Markov Chain\nStationary Distribution Samples")
for c in range(C):
    ax[0].plot(theta[c,:burn], label="$\\theta$ Chain "+str(c))
    ax[1].plot(tau[c,:burn], label="$\\tau$ Chain "+str(c))
    ax[2].plot(np.arange(burn, demo, dtype=int), theta[c,burn:demo], label="$\\theta$ Chain "+str(c))
    ax[3].plot(np.arange(burn, demo, dtype=int), tau[c,burn:demo], label="$\\tau$ Chain "+str(c))
ax[0].legend(); ax[1].legend(); ax[2].legend(); ax[3].legend();

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot for theta
axes[0].hist(posterior_theta, bins=30, density=True, alpha=0.6, color='b')
axes[0].axvline(interval_theta[0], color='red', linestyle='dashed', label="2.5th percentile")
axes[0].axvline(interval_theta[1], color='red', linestyle='dashed', label="97.5th percentile")
axes[0].set_title(r'Posterior Distribution of $\theta$')
axes[0].set_xlabel(r'$\theta$')
axes[0].set_ylabel('Density')
axes[0].legend()

# Plot for tau
axes[1].hist(posterior_tau, bins=30, density=True, alpha=0.6, color='r')
axes[1].axvline(interval_tau[0], color='blue', linestyle='dashed', label="2.5th percentile")
axes[1].axvline(interval_tau[1], color='blue', linestyle='dashed', label="97.5th percentile")
axes[1].set_title(r'Posterior Distribution of $\tau$')
axes[1].set_xlabel(r'$\tau$')
axes[1].set_ylabel('Density')
axes[1].legend()

plt.show()
```

#Q2 Response
We are given $p(\theta) = N(\theta_0, \tau_0)$ and $p(\tau) \propto \tau^{\frac{\alpha}{2}-1} e^{- \frac{\lambda}{2} \tau}$ \
First we find $p(\tau|\theta,x)$: \

$p(\theta)\propto \frac{\sqrt{\tau}}{\sqrt{2\pi}}exp(\frac{-\tau(\theta-\theta_0)^2}{2})$

$$
p(\theta | \tau, x) \propto p(x | \theta, \tau) p(\theta).
$$

$$
p(x | \theta, \tau) \propto \tau^{n/2} \exp\left( -\frac{\tau}{2} \sum_{i=1}^{n} (x_i - \theta)^2 \right).
$$

$$
p(\tau | \theta, \mathbf{x}) \propto \tau^{\frac{n}{2}} \exp\left( -\frac{\tau}{2} \sum_{i=1}^{n} (x_i - \theta)^2 \right) \cdot \tau^{\frac{\alpha}{2} - 1} \exp\left( -\frac{\lambda}{2} \tau \right)
$$

$$
p(\tau|\theta, \mathbf{x}) \sim \text{Gamma} \left( \frac{n + \alpha}{2}, \frac{\sum_{i=1}^{n} (x_i - \theta)^2 + \lambda}{2} \right)
$$

Now for $p(\theta|\tau,x)$: \

$$
p(\theta|\tau,x) \propto exp(-\frac{\tau}{2} \sum^n_{i=1}(x_i - \theta)^2 - \frac{\tau_0}{2} \sum^n_{i=1}(\theta - \theta_0)^2)
$$

which is simply:
$$
p(\theta|\tau,x) \sim N(\frac{n\tau\bar{x} + \theta_0 \tau_0}{n\tau + \tau_0}, n\tau + \tau_0)
$$

part 2:  
$\alpha$ here controls the weight we give to our prior. As alpha increase, we would require more data to change the posterior by similar amount. $\lambda$ here mainly controls the dispersion of $\tau$, where larger $\lambda$ means less variance on $\tau$(Since var of gamma is $\frac{\alpha}{\lambda^2}$). It also contribute to the mean of $\tau$, increasing it will decrease the mean. \
the "number of data points" contributed through the likelihood can be evaluated as $\frac{\tau_0}{\tau}$. Combining this with the effect of $\lambda$ and $\alpha$ on $\tau$ give us an understanding of how changing these two parameters change the "number of data points" contributed. We may say that the prior contributes fewer data points as $\alpha$ increase and $\lambda$ decrease.

part3:
From the plots, we see that the posterior distribution moves further away from our preset prior as we increase $\alpha$, and the confidence interval gets smaller as we increase $\lambda$. Which matches our analysis from part2: decreasing $\alpha$ will move our posterior toward our prior, and increasing $\lambda$ makes the confidence inverval wider.

```python
#Q2 part3

def gibbs_sampler(alpha, lam, theta_0, tau_0):
    # lam is lambda

    # loc is mean and scale is standard deviation
    n = 30; x = stats.norm(loc=0, scale=1).rvs(size=n)
    C,G = 2,10000  # chains, Gibbs samples
    theta,tau = np.zeros([C,G]),np.zeros([C,G])
    theta[:,0] = 1000  # initialization

    for c in range(C):
        for g in range(1,G):
            shape = (n + alpha) /2
            scale = 2/(np.sum((x - theta[c, g - 1])**2) + lam)
            tau[c,g] = stats.gamma(a=shape, scale=scale).rvs()

            #update theta
            theta[c, g] = stats.norm((tau[c, g] * np.sum(x) + tau_0 * theta_0) / (tau[c, g] * n + tau_0), scale=np.sqrt(1/(tau[c, g] * n + tau_0))).rvs()

    return theta, tau

test_size = 9
alpha_values = [0.1, 1, 10, 0.1, 1, 10, 0.1, 1, 10]
lambda_values = [0.1, 0.1, 0.1, 10, 10, 10, 100, 100, 100]

theta_0 = 0
tau_0 = 1

all_posteriors_theta = []
all_posteriors_tau = []

# Test over all combinations of alpha and lambda values
for i in range(test_size):
        # Run the Gibbs sampler for each combination
        theta, tau = gibbs_sampler(alpha_values[i], lambda_values[i], theta_0, tau_0)

        # Burn-in and demo samples
        burn = 20
        demo = 120
        posterior_theta = theta[:, burn:].flatten()
        posterior_tau = tau[:, burn:].flatten()

        # Store the results for each setting
        all_posteriors_theta.append(posterior_theta)
        all_posteriors_tau.append(posterior_tau)

        # Compute 95% credible intervals
        interval_theta = np.percentile(posterior_theta, [2.5, 97.5])
        interval_tau = np.percentile(posterior_tau, [2.5, 97.5])

        print(f"Alpha: {alpha_values[i]}, Lambda: {lambda_values[i]}")
        print("95% CI for theta:", interval_theta)
        print("95% CI for tau:", interval_tau)

        # Plotting posterior distributions with 95% CI
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        # Plot for theta
        axes[0].hist(posterior_theta, bins=30, density=True, alpha=0.6, color='b')
        axes[0].axvline(interval_theta[0], color='red', linestyle='dashed', label="2.5th percentile")
        axes[0].axvline(interval_theta[1], color='red', linestyle='dashed', label="97.5th percentile")
        axes[0].set_title(f'Posterior Distribution of $\\theta$ (Alpha={alpha_values[i]}, Lambda={lambda_values[i]})')
        axes[0].set_xlabel(r'$\theta$')
        axes[0].set_ylabel('Density')
        axes[0].legend()

        # Plot for tau
        axes[1].hist(posterior_tau, bins=30, density=True, alpha=0.6, color='r')
        axes[1].axvline(interval_tau[0], color='blue', linestyle='dashed', label="2.5th percentile")
        axes[1].axvline(interval_tau[1], color='blue', linestyle='dashed', label="97.5th percentile")
        axes[1].set_title(f'Posterior Distribution of $\\tau$ (Alpha={alpha_values[i]}, Lambda={lambda_values[i]})')
        axes[1].set_xlabel(r'$\tau$')
        axes[1].set_ylabel('Density')
        axes[1].legend()

        plt.show()
```

$$
P(\theta | x) \propto P(x | \theta) \cdot P(\theta)
$$

In principle, one could identify the exact distribution of the posterior by looking at the joint distribution, combining the likelihood and the prior, and normalizing the result to ensure the posterior sums (or integrates) to one. As that previous line is from:

$$
P(\theta | x) = \frac{P(x| \theta) \cdot P(\theta)}{P(x)}
$$

And being simplified by stating p(x) is a constant.

Drawing important information without looking at the constant there just to ensure everything sums to one.
